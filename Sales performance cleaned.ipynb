```python
# Step 2: Remove unnecessary columns (Example: dropping unnamed index columns)
unnecessary_columns = [col for col in data.columns if "unnamed" in col.lower()]
data.drop(columns=unnecessary_columns, inplace=True)                  ##there is no unnamed column
```

```python
# Step 4: Handle inconsistent formatting (e.g., trimming whitespace, fixing case sensitivity)
data.columns = data.columns.str.strip()  # Remove leading/trailing spaces in column names
```

```python
import re
# Function to clean column names
def clean_column_names(df):
    df.columns = [re.sub(r'[^a-zA-Z0-9 ]', '', col).strip().replace("_","-") for col in df.columns]
    
    return df

# Function to remove currency symbols $ from specific columns
def remove_currency_symbols(df, columns):
    for col in columns:
         df[col] = df[col].astype(str).replace(r'[\$,()]', '', regex=True)  # Correct regex pattern
         df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)  # Convert to integer
    return df

# Apply cleaning functions
data = clean_column_names(data)
currency_columns = ['Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts',
       'Sales', 'COGS', 'Profit']
data = remove_currency_symbols(data, currency_columns)

# Display cleaned column names
print("Cleaned Column Names:")
print(data.columns)
```
